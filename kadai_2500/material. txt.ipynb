```python
from google.colab import files
import pandas as pd

# ファイルアップロード
uploaded = files.upload()
print("アップロードされたファイル:", uploaded.keys())

# ファイル名を指定（実際のアップロード名に一致させる）
conditions_file = "condition.xlsx"
materials_file = "materials_data_with_density.xlsx"

# ファイルがアップロードされているか確認
if conditions_file not in uploaded.keys():
    raise FileNotFoundError(f"{conditions_file} がアップロードされていません。")
if materials_file not in uploaded.keys():
    raise FileNotFoundError(f"{materials_file} がアップロードされていません。")

# シート名の確認
try:
    print("条件ファイルのシート名:", pd.ExcelFile(conditions_file).sheet_names)
    print("材料データファイルのシート名:", pd.ExcelFile(materials_file).sheet_names)
except Exception as e:
    print("シート名の確認中にエラーが発生しました:", e)
    raise

# Excelファイルの読み込み
try:
    conditions_df = pd.read_excel(conditions_file, sheet_name="条件")  # シート名を確認して修正
    materials_df = pd.read_excel(materials_file, sheet_name="材料データ")  # シート名を確認して修正
except Exception as e:
    print("Excelファイルの読み込み中にエラーが発生しました:", e)
    raise

# 条件を辞書に変換
try:
    conditions = conditions_df.set_index("条件項目")["値"].to_dict()
    print("条件:", conditions)
    print("材料データ:", materials_df.head())
except Exception as e:
    print("条件または材料データの処理中にエラーが発生しました:", e)
    raise
```
